```python
#Feature Selection and Data Preprocessing
```

```python
# Drop unnecessary columns
df_filtered = df.drop(columns=['player_dismissed', 'dismissal_kind', 'fielder', 'non_striker'])

# Convert categorical features to 'category' type and assign numerical codes
for col in ['batting_team', 'bowling_team', 'batsman', 'bowler']:
    df_filtered[col] = df_filtered[col].astype('category').cat.codes

# Display the first few rows after encoding
df_filtered.head()
```

```python
df_filtered
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Generate the correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_filtered.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlations")
plt.show()
```

This heatmap illustrates the correlation coefficients among various cricket match features. Darker reds signify strong positive correlations, while darker blues indicate strong negative correlations.

```python
from sklearn.model_selection import train_test_split

# Define features (X) and target variable (y)
X = df_filtered.drop(columns=['total_runs'])
y = df_filtered['total_runs']

# Split into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_test
```

```python
len(X_train)
```

```python
from sklearn.linear_model import LinearRegression

# Initialize and train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions on test set
y_pred = model.predict(X_test)


```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R² Score: {r2}")
```

```python
# Scatter plot of actual vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], '--', color='red')  # Ideal line
plt.xlabel("Actual Runs")
plt.ylabel("Predicted Runs")
plt.title("Actual vs. Predicted Runs (Regression Model)")
plt.show()
```
